```python
results = model.predict(source='/content/dreamcatcher.mp4', save=True)
```

```python
import cv2

# 원본 영상 열기
cap = cv2.VideoCapture('dreamcatcher.mp4')
frame_width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
frame_height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
fps = int(cap.get(cv2.CAP_PROP_FPS))

# 감지된 객체 정보 추출
detected_objects = []
for result in results:
    frame_objects = []  # 현재 프레임의 객체 리스트
    for box in result.boxes.xyxy:  # 각 객체의 경계 상자 좌표 (x1, y1, x2, y2)
        x1, y1, x2, y2 = map(int, box)
        frame_objects.append((x1, y1, x2, y2))
    detected_objects.append(frame_objects)

# 객체별 VideoWriter 생성
num_objects = len(detected_objects[0])  # 첫 번째 프레임의 객체 수 기준
writers = [cv2.VideoWriter(f'object_{i + 1}.mp4', cv2.VideoWriter_fourcc(*'mp4v'), fps, (100, 100)) for i in range(num_objects)]

frame_count = 0

while cap.isOpened():
    ret, frame = cap.read()
    if not ret or frame_count >= len(detected_objects):
        break

    # 현재 프레임에서 객체 좌표 가져오기
    objects = detected_objects[frame_count]

    for i, (x1, y1, x2, y2) in enumerate(objects):
        # 객체별로 프레임 자르기
        cropped_frame = frame[y1:y2, x1:x2]
        cropped_frame = cv2.resize(cropped_frame, (100, 100))  # 원하는 크기로 조정

        # 객체별 영상에 프레임 추가
        writers[i].write(cropped_frame)

    frame_count += 1

# 자원 해제
cap.release()
for writer in writers:
    writer.release()

# 저장된 영상 파일 다운로드
from google.colab import files
for i in range(num_objects):
    files.download(f'object_{i + 1}.mp4')
```
